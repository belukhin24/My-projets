```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats as st
import math as mth
```

```python
df = pd.DataFrame([['клики',25130, 23750], ['установки' , 1650, 1345], ['покупки', 345, 285]], columns =['event_name', 'group_a', 'group_b']) 
df
```

```python
sum_users = df[:1]
sum_users
```

```python
alpha = 0.05

def z_func(group1, group2):
    for i in df.index:
        # пропорция успехов в первой группе:                        
        p1 = df[group1][i] / sum_users[group1]
        # пропорция успехов в второй группе:
        p2 = df[group2][i] / sum_users[group2]
        # пропорция успехов в комбинированном датасете:
        p_combined = (
            (df[group1][i] + df[group2][i]) 
            / (sum_users[group1] + sum_users[group2])
        )
        # разница пропорций в датасетах
        difference = p1 - p2
        # считаем статистику в ст.отклонениях стандартного нормального распределения
        z_value =(
            difference 
            / mth.sqrt(p_combined * (1 - p_combined) 
                       * (1/sum_users[group1] 
                          + 1/sum_users[group2]))
        )
        # задаем стандартное нормальное распределение (среднее 0, ст.отклонение 1)
        distr = st.norm(0, 1)

        p_value = (1 - distr.cdf(abs(z_value))) * 2

        print('В результате сравнения результатов групп {} и {} по мероприятию {:*^17} p_value равно: {}'.format(group1, 
                                                                                                                 group2,
                                                                                                                 df['event_name'][i], 
                                                                                                                 p_value))

        if p_value < alpha:
            print('Основываясь на полученный p value отвергаю нулевую гипотезу: между долями есть значимая разница')
        else:
            print('Основываясь на полученный p value, не получилось отвергнуть нулевую гипотезу, нет оснований считать доли разными')
        print()
  
```

```python
z_func('group_a', 'group_b')
```
